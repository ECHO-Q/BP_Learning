```python
from tensorflow.examples.tutorials.mnist import input_data
import tensorflow as tf
```

```python
# one_hot向量除了某一位数字是1以外，其余维度数字都是0
# labels:0-9
mnist=input_data.read_data_sets("MNIST_data",one_hot=True) 

# 定义每层的节点数量
input_size=784
output_size=10
hidden_size=50
batch_size=100
n_batch=mnist.train.num_examples//batch_size  #一共多少批次

# input layer
x_data=tf.placeholder(tf.float32,[None,input_size])

# hidden layer
hidden_weight=tf.Variable(tf.truncated_normal([input_size,hidden_size]))
hidden_bias=tf.Variable(tf.truncated_normal([hidden_size]))
hidden_layer=tf.add(tf.matmul(x_data,hidden_weight),hidden_bias)
hidden_layer=tf.nn.sigmoid(hidden_layer)

# output layer
output_weight=tf.Variable(tf.truncated_normal([hidden_size,output_size]))
output_bias=tf.Variable(tf.zer([output_size]))
output_layer=tf.add(tf.matmul(hidden_layer,output_weight),output_bias)
output_layer=tf.nn.sigmoid(output_layer)

# output label
y_data=tf.placeholder(tf.float32,[None,output_size])

# loss function
loss=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels=y_data,logits=output_layer))

# optimizer
train_step=tf.train.AdamOptimizer().minimize(loss)

# accuracy
correct_prediction=tf.equal(tf.argmax(y_data,1),tf.argmax(output_layer,1))
accuracy=tf.reduce_mean(tf.cast(correct_prediction,tf.float32))

# training
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for epoch in range(50):
        for batch in range(n_batch):
            batch_xs,batch_ys=mnist.train.next_batch(batch_size)
            sess.run(train_step,feed_dict={x_data:batch_xs,y_data:batch_ys})
        
        acc=sess.run(accuracy,feed_dict={x_data:mnist.test.images,y_data:mnist.test.labels})
        print("Iter"+str(epoch)+",Testing Accuracy "+str(acc))    
    
```
